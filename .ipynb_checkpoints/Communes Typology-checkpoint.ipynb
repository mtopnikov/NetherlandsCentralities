```python
def plot_types(typology, year):
    typology.plot.scatter(x = 'flow', y = 'centrality', color = colors['peach'], figsize = (7,7))

    plt.vlines(typology.flow.mean(), ymin = 0, ymax = 1, color = 'r')
    plt.vlines(typology.flow.mean() + 3 * typology.flow.std(),
               ymin = 0, ymax = 1,
               color = 'r',
               linestyles = 'dashed')
    # plt.vlines(typology.flow.mean() - 2 * typology.flow.std(), ymin = 0, ymax = 1, color = 'r', linestyles = 'dashed')
    plt.hlines(typology.centrality.mean(), xmin = 0, xmax = 1)
    plt.hlines(typology.centrality.mean() + 3 * typology.centrality.std(), xmin = 0, xmax = 1, linestyles = 'dashed')

    plt.text(typology.flow.mean() +.01,
             .97, '$\\bar x$', color = "r", fontsize = 13)
    plt.text(typology.flow.mean()+ 3 * typology.flow.std()+.01,
             .97, '$\\bar x + 3\sigma$', color = "r",  fontsize = 13)

    plt.text(1, typology.centrality.mean() +.01,
             '$\\bar y$', color = "k", fontsize = 13, ha = 'right')
    plt.text(1, typology.centrality.mean()+ 3 * typology.centrality.std()+.01,
             '$\\bar y + 3\sigma$', color = "k",  fontsize = 13,  ha = 'right')

    plt.grid()

    plt.xlim(-.01,1.01)
    plt.ylim(-.01,1.01)

    plt.xticks(fontproperties = regluar_font)
    plt.yticks(fontproperties = regluar_font)

    plt.xlabel('Средняя интенсивность поездки',fontproperties = medium_font)
    plt.ylabel('Центральность муниципалитета',fontproperties = medium_font)
    plt.tight_layout()
    plt.savefig(f'analysis/typology/plots/typology_{year}.png', dpi = 192)
    plt.close('all')
```

```python
def types_detect(x, flow_mean, centrality_mean):
    if x['flow'] < flow_mean and x['centrality'] < centrality_mean:
        return 'low_low'
    if x['flow'] < flow_mean and x['centrality'] > centrality_mean:
        return 'high_low'
    if x['flow'] > flow_mean and x['centrality'] > centrality_mean:
        return 'high_high'
    if x['flow'] > flow_mean and x['centrality'] < centrality_mean:
        return 'low_high'
```

```python
typology['type'] = typology.apply(lambda x: types_detect(x, typology['flow'].mean(), typology['centrality'].mean()), axis = 1)
```

```python
ratio = typology['flow'].mean()/  typology['centrality'].mean()
```

```python
typology.loc[
    typology.type == 'low_low', 'type'
] = typology.loc[typology.type == 'low_low'].apply(
    lambda x: 'lower_' + types_detect(x,
                                      typology[typology.type == 'low_low'].flow.mean(), 
                                      typology[typology.type == 'low_low'].centrality.mean()),
    axis = 1
)
```
