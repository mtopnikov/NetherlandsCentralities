```python
with open('Variables/mappers/values_ru_map_v2.json') as infile:
    values_map = json.load(infile)
```

```python
colnames_map = {    
    "wogem": "geo_commune",
    "vertgem": "geo_departure",
    "aankgem": "geo_arrival",
    "motief": "motiv_motivation",
    "motief": "motiv_motivation",
    "motiefv": "motiv_motivation"
}
```

```python
motivation_map = {   
    'OViN': {
        1 : 'Работа',
        2 : 'Рабочая встреча',
        3 : 'Транспортная сфера работы',
        4 : 'Перевозка людей',
        5 : 'Перевозка товаров',
        6 : 'Образование',
        7 : 'Покупки',
        8 : 'Посещение друзей и родственников',
        9 : 'Туризм и рекреация',
        10 : 'Спорт',
        11 : 'Другие виды досуга',
        12 : 'Услуги',
        13 : 'Другое'
    },
    'MON' : {
        0 : 'Не указан',
        2 : 'Работа',
        3 : 'Рабочая встреча',
        4 : 'Транспортная сфера работы',
        5 : 'Перевозка людей',
        6 : 'Перевозка товаров',
        7 : 'Образование',
        8 : 'Покупки',
        9 : 'Посещение друзей и родственников',
        10 : 'Туризм и рекреация',
        11 : 'Спорт',
        12 : 'Другие виды досуга',
        13 : 'Услуги',
        14 : 'Совместно с другим человеком',
        15 : 'Другое'
    }
}
```

```python
for data in iterator:
    valuesMap = values_map[data]
    l_dir = iterator[data]['list_dir']
    root_path = iterator[data]['path']
    for l in tqdm(l_dir):
        df, meta = pyreadstat.read_sav(root_path + l)
        df.columns = [i.lower() for i in df.columns]
        
        new_labels = {k.lower() : v for k, v in meta.variable_value_labels.items()}
        

        df = df[
            set(colnames_map.keys()).intersection(set(df.columns))
        ].dropna(subset = ['vertgem', 'aankgem'])

            
        
        df = df.apply(
            lambda column: column.map(new_labels['wogem']) if column.name in ['wogem', 'vertgem', 'aankgem'] else column
        ).apply(
            lambda column: column.map(communes_mapper) if column.name in ['wogem', 'vertgem', 'aankgem'] else column
        )
        df.dropna(subset = ['vertgem', 'aankgem'], inplace = True)
        
        df.rename(columns = colnames_map, inplace = True)
        df['motiv_motivation'] = df['motiv_motivation'].astype(int).map(motivation_map[data])
        df['uid_person'] = pd.Series([i for i in range(len(df))])        
        
        
        year = re.findall('\d+', l)[0]
        df.to_csv(f'translated_data_v3/data_{year}.csv', index = None)
```

```python
for year in range(2010, 2018):
    try:
        df = pd.read_spss(f'OViN_dat/OViN{year}_Databestand.sav')
    except:
        df = pd.read_spss(f'OViN_dat/OViN{year}_Databestand_revisie.sav')
    df.columns = [i.lower() for i in df.columns]
    
    df.rename(columns = {'vertgem' : 'geo_departure', 'aankgem' : 'geo_arrival'}, inplace = True)
    df['uid_person'] = pd.Series([i for i in range(len(df))])
    df.dropna(subset = ['geo_departure', 'geo_arrival'], inplace = True)
    matrix = get_corrMatrix(df)
    matrix = matrix[matrix.local_1 != matrix.local_2]
    matrix['flow'] = matrix['flow'] / matrix['flow'].sum()
    
    coords = gpd.read_file(f'GeementenShapes/communes_{year}.shp').to_crs({'init' : 'epsg:4326'})
    coords.set_index('statnaam', inplace = True)
    coords['geometry'] = coords['geometry'].centroid
    coords['coords'] = coords['geometry'].apply(lambda x: [x.x, x.y]) 
    dict_coords = coords['coords'].to_dict()
    
    matrix['l1_coords'] = matrix['local_1'].map(dict_coords)
    matrix['l2_coords'] = matrix['local_2'].map(dict_coords)

    matrix.dropna(subset = ['l1_coords', 'l2_coords'], inplace = True)
    
    gdf_matrix = gpd.GeoDataFrame(
        matrix,
        crs = {'init' : 'epsg:4326'},
        geometry = [LineString([c1, c2]) for c1, c2  in zip(matrix.l1_coords, matrix.l2_coords)]
    )
    gdf_matrix.drop(columns = ['l1_coords', 'l2_coords'], inplace = True)
    
    gdf_matrix.to_file(f'REMAKE/graphs/{year}_graph.gpkg', driver = 'GPKG')
```
